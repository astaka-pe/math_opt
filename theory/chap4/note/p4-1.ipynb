```python
import networkx as nx
import matplotlib.pyplot as plt
```

# 無向グラフ

```python
G = nx.Graph()
vlist = [1, 2, 3, 4]
elist = [(1, 2), (1, 3), (2, 3), (2, 4), (3, 4)]
G.add_nodes_from(vlist)
G.add_edges_from(elist)
nx.draw_networkx(G, node_color="lightgray", edge_color="green", node_size=400)
plt.axis("off")
plt.show()
```

```python
print(G.nodes(), G.number_of_nodes())
print(list(nx.all_neighbors(G, 1)))
print("隣接行列")
print(nx.adjacency_matrix(G).todense())
print("接続行列")
print(nx.incidence_matrix(G).todense())
```

# 有向グラフ

```python
DG = nx.DiGraph()
vlist = [1, 2, 3, 4]
elist = [(1, 1), (1, 3), (2, 3), (2, 4), (3, 4)]
DG.add_nodes_from(vlist)
DG.add_edges_from(elist)
nx.draw_networkx(DG, node_color="lightgray", edge_color="green", node_size=400)
plt.axis("off")
plt.show()
```

```python
print("隣接行列")
print(nx.adjacency_matrix(DG).todense())
print("接続行列")
print(nx.incidence_matrix(DG, oriented=True).todense())
```

# 完全グラフ

```python
G = nx.complete_graph(5)
p = nx.spring_layout(G, iterations=100)
#p = nx.circular_layout(G)
nx.draw_networkx(G, pos=p, node_color="lightgray", edge_color="green", node_size=300)
plt.axis=("off")
plt.show()
```

# 完全2部グラフ

```python
m, n = 3, 4
G = nx.complete_bipartite_graph(m, n)
p = {}
for i in range(m):
    p[i] = (0, i)
for i in range(n):
    p[m+i] = (1, i)
nx.draw_networkx(G, pos=p, node_color="lightgray", edge_color="green", node_size=300)
plt.axis=("off")
plt.show()
```
